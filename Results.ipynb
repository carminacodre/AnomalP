### Averages for each family

```python
df = pd.read_csv("results_proteins/res_fam_prot_angles.csv")
del df['Iteration']
f1 = []
acc = []
df['F1'] = 2.0 * df['Prec'] * df['Recall'] / (df['Prec'] + df['Recall'])
df['Acc'] = (df['TP'] + df['TN']) / (df['TP'] + df['TN'] + df['FP'] + df['FN'])
df.head()
```

```python
print("Precision")
avg_fam = df.groupby(['Superfamily'])['Prec'].agg(['mean', 'count', 'std'])
avg_fam
ci = []
for i in avg_fam.index:
    m, c, s = avg_fam.loc[i]
    ci.append(1.96*s/math.sqrt(c))
avg_fam['ci95'] = ci
avg_fam
```
