# Results statistics

```python
import pandas as pd
import math
```

### Averages for each family

```python
df = pd.read_csv("/Users/carminacodre/Desktop/results_proteins/res_fam_conf_sa.csv")
del df['Iteration']
del df['TP']
del df['TN']
del df['FP']
del df['FN']
f1 = []
for i in df.index:
    f, prec, rec, spec, auc = df.loc[i]
    f1.append(2.0 * prec * rec / (prec + rec))
df['F1'] = f1
df
```

```python
print("Precision")
avg_fam = df.groupby(['Superfamily'])['Prec'].agg(['mean', 'count', 'std'])
avg_fam
ci = []
for i in avg_fam.index:
    m, c, s = avg_fam.loc[i]
    ci.append(1.96*s/math.sqrt(c))
avg_fam['ci95'] = ci
avg_fam
```

```python
print("Recall")
avg_fam = df.groupby(['Superfamily'])['Recall'].agg(['mean', 'count', 'std'])
avg_fam
ci = []
for i in avg_fam.index:
    m, c, s = avg_fam.loc[i]
    ci.append(1.96*s/math.sqrt(c))
avg_fam['ci95'] = ci
avg_fam
```
